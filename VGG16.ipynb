```
from tensorflow.keras.applications.vgg16 import VGG16
```

```
from tensorflow.keras.utils import plot_model
```

```
model = VGG16()
```

```
plot_model(model,to_file='vgg.png')
```

```
model.summary()
```

```
target = (224,224)
```

```
from tensorflow.keras.preprocessing.image import load_img
```

```
cat = load_img('/content/catto.jpg',target_size=target)
dog = load_img('/content/doggo.jpg',target_size=target)
jfl = load_img('/content/vgg.png',target_size=target)
```

```
from tensorflow.keras.preprocessing.image import img_to_array
```

```
cat = img_to_array(cat)
dog = img_to_array(dog)
jfl = img_to_array(jfl)
```

```
cat = cat.reshape((1,cat.shape[0],cat.shape[1],cat.shape[2]))
dog = dog.reshape((1,dog.shape[0],dog.shape[1],dog.shape[2]))
jfl = jfl.reshape((1,jfl.shape[0],jfl.shape[1],jfl.shape[2]))
```

```
from tensorflow.keras.applications.vgg16 import preprocess_input
```

```
cat = preprocess_input(cat)
dog = preprocess_input(dog)
jfl = preprocess_input(jfl)
```

```
catpred = model.predict(cat)
dogpred = model.predict(dog)
jflpred = model.predict(jfl)
```

```
from tensorflow.keras.applications.vgg16 import decode_predictions
```

```
import matplotlib.pyplot as plt
```

```
plt.imshow(load_img('/content/catto.jpg'))
decode_predictions(catpred)
```

```
plt.imshow(load_img('/content/doggo.jpg'))
decode_predictions(dogpred)
```

```
plt.imshow(load_img('/content/vgg.png'))
decode_predictions(jflpred)
```
