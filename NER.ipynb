```python
data = pd.read_csv("/content/Named-Entity-Recognition/ner_dataset.csv", encoding='latin1')
data.head(30)
```

```python
data = data.fillna(method='ffill')
data.head(30)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data['Sentence #'] = LabelEncoder().fit_transform(data['Sentence #'])
data.head(30)
```

```python
data.rename(columns={'Sentence #':'sentence_id', 'Word':'words', 'Tag':'labels'}, inplace=True)
data['labels'] = data['labels'].str.upper()
data.head()
```

```python
x = data[['sentence_id', 'words']]
y = data['labels']
print(x,y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

```python
train_data = pd.DataFrame({'sentence_id':x_train['sentence_id'], 'words':x_train['words'], 'labels': y_train})
test_data = pd.DataFrame({'sentence_id':x_test['sentence_id'], 'words':x_test['words'], 'labels': y_test})
```

```python
train_data
```

```python
# Model Training
```

```python
from simpletransformers.ner import NERModel, NERArgs
```

```python
labels = data['labels'].unique().tolist()
labels
```

```python
args = NERArgs()
args.num_train_epochs = 1
args.learning_rate = 1e-4
args.overwrite_output_dir = True
args.train_batch_size = 32
args.test_batch_size = 32
args.fp16 = True
args.use_cuda = True
```

```python
model = NERModel('bert', 'bert-base-cased', labels = labels, args = args)
```

```python
model.train_model(train_data,eval_data = test_data,acc=accuracy_score)
```

```python
result, model_outputs, preds_list = model.eval_model(test_data)
result
```

```python
prediction, model_output = model.predict(["Is Albert living in California"])
prediction
```
